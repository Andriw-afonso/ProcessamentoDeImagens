##   Normalizando os dados

### Transformando a imagem em uma array numpy

```python
# Carregando a imagem nor formato em array numpy
pixels = asarray(image)
pixels
```

```python
# Verificando os tipos de dados e valores maximos e minimos
print('Data types: {}'.format(pixels.dtype))
print('Min: {}, Max: {}'.format(pixels.min(),pixels.max()))
```

### Aplicando a normalizao

```python
# Transformando os dados em numeros reais
pixels = pixels.astype('float32')

# Normalizando
pixels /= pixels.max()

# Visualizando os tipos dos dados e seus valores minimos e maximos
print('Data types: {}'.format(pixels.dtype))
print('Min: {}, Max: {}'.format(pixels.min(),pixels.max()))
```
